```python
# Lets restrict our dataset to only MSAs that have data from 2000 onward
zhvi_clean = zhvi[zhvi['2000-01-31'].notna()]
```

```python
zhvi_clean
```

```python
# Identify the columns that are not dates
top_100_sizerank = zhvi_clean[zhvi_clean['SizeRank'] < 113]
id_vars = ['RegionID', 'SizeRank', 'RegionName', 'RegionType', 'StateName']

# Use pd.melt to convert the DataFrame from wide to long format
df_long = pd.melt(top_100_sizerank, id_vars=id_vars, var_name='Date', value_name='Value')

# Convert the Date column to datetime type
df_long['Date'] = pd.to_datetime(df_long['Date'])

# Reset the index
df_long.reset_index(drop=True, inplace=True)

# Create a pivot table with Date as index and other columns as headers
df_pivot = df_long.pivot_table(index='Date', columns=['RegionID'], values='Value')
df_pivot = df_pivot.drop(102001, axis=1)
df_pivot = df_pivot.fillna(method='ffill')

# Display the first few rows of the pivoted DataFrame
print("Pivoted DataFrame:")
print(df_pivot.head())

df_pivot.to_csv('../data/processed/zhvi_top100_ts.csv')
```
